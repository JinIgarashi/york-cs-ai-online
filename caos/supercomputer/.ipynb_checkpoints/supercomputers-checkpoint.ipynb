# Supercomputers plots

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import textwrap
import matplotlib.patches as mpatches
from matplotlib.path import Path
import matplotlib.ticker as ticker
```

```python
path_excel = r'Top500 Nov 2023.xlsx'
sheet_name = '62'
```

```python
df = pd.read_excel(
    path_excel, 
    sheet_name=sheet_name,
)
print(df)
```

```python
data = {}
for year, group in df.groupby('Year'):
    data[year] = group['Rpeak [TFlop/s]'].values

# ボックスプロットを描画する
plt.figure(figsize=(10, 6))
plt.boxplot(data.values(), labels=data.keys())
plt.title('Boxplot of Rpeak [TFlop/s] by Year')
plt.xlabel('Year')
plt.ylabel('Rpeak [TFlop/s]')
plt.xticks(rotation=45)  # X軸のラベルを回転させる
plt.tight_layout()
plt.show()
```

```python
# 各年度から最大のRpeak [TFlop/s]を持つ行を抽出
max_rpeak_rows = df.loc[df.groupby('Year')['Rpeak [TFlop/s]'].idxmax()]

# プロット用のデータを準備
x_values = max_rpeak_rows['Year']
y_values = max_rpeak_rows['Rpeak [TFlop/s]']
names = max_rpeak_rows['Name']
manufacturers = max_rpeak_rows['Manufacturer']

# ポイントをプロット
plt.figure(figsize=(10, 6))
plt.scatter(x_values, y_values, marker='o')

# ラベルを追加
for name, manufacturer, x, y in zip(names, manufacturers, x_values, y_values):
    if pd.notna(name):
        label_text = f"{name}\n({manufacturer})"  # Nameの下にManufacturerを括弧で括って表示
    else:
        label_text = f"{manufacturer}"  # Nameが未指定の場合はManufacturerの名前のみを表示
    bbox_props = dict(boxstyle="round,pad=0.3", fc="white", ec="black", lw=0.5, alpha=1.0)  # 背景の設定
    y_adjust = 100000
    if x == 2016:
        y_adjust = 200000
    elif x == 2021:
        y_adjust = -200000
    plt.text(x, y + y_adjust, label_text, ha='center', va='center', bbox=bbox_props)  # ラベルを中央揃えに設定

# ポイント同士を線で結ぶ
plt.plot(x_values, y_values, linestyle='-', color='b')

# # Y軸の範囲を設定
plt.ylim(0, 1800000)

# Y軸のフォーマットをmillionに変更
formatter = ticker.FuncFormatter(lambda x, _: '{:.1f}'.format(x / 1e6))
plt.gca().yaxis.set_major_formatter(formatter)

# ラベルを設定
# plt.title('Yearly supercomputer performance')
plt.xlabel('Year')
plt.ylabel('Max Rpeak [TerraFlop/s]')
plt.xticks(x_values, rotation=90)  # X軸のラベルを回転させる
plt.tight_layout()
plt.savefig('supercomputer_max_rpeak_plot.png', dpi=600)
plt.show()


```

```python
# 各年度から最大のRpeak [TFlop/s]とTotal Coresを持つ行を抽出
max_rpeak_rows = df.loc[df.groupby('Year')['Rpeak [TFlop/s]'].idxmax()]
max_cores_rows = df.loc[df.groupby('Year')['Total Cores'].idxmax()]

# プロット用のデータを準備
x_values = max_rpeak_rows['Year']
y_values_rpeak = max_rpeak_rows['Rpeak [TFlop/s]']
y_values_cores = max_cores_rows['Total Cores']
names = max_rpeak_rows['Name']
manufacturers = max_rpeak_rows['Manufacturer']

# ポイントをプロット
fig, ax1 = plt.subplots(figsize=(10, 6))

# Rpeak [TFlop/s]をプロット
ax1.scatter(x_values, y_values_rpeak, marker='o', label='Max Rpeak [TFlop/s]')
ax1.plot(x_values, y_values_rpeak, linestyle='-', color='b')
ax1.set_ylabel('Max Rpeak [TFlop/s]')  # 左側のY軸のラベルを設定

# Y軸の範囲を設定
ax1.set_ylim(0, 1800000)

# Y軸のフォーマットをmillionに変更
formatter = ticker.FuncFormatter(lambda x, _: '{:.1f}'.format(x / 1e6))
ax1.yaxis.set_major_formatter(formatter)

# Total Coresを2番目のY軸にプロット
ax2 = ax1.twinx()
ax2.scatter(x_values, y_values_cores, marker='s', color='r', label='Total Cores')
ax2.plot(x_values, y_values_cores, linestyle='--', color='r')
ax2.set_ylabel('Total Cores')  # 右側のY軸のラベルを設定

# ラベルを追加
for name, manufacturer, x, y_rpeak, y_cores in zip(names, manufacturers, x_values, y_values_rpeak, y_values_cores):
    if pd.notna(name):
        label_text = f"{name}\n({manufacturer})"  # Nameの下にManufacturerを括弧で括って表示
    else:
        label_text = f"{manufacturer}"  # Nameが未指定の場合はManufacturerの名前のみを表示
    bbox_props = dict(boxstyle="round,pad=0.3", fc="white", ec="black", lw=0.5, alpha=1.0)  # 背景の設定
    y_adjust = 100000
    if x == 2016:
        y_adjust = 200000
    elif x == 2021:
        y_adjust = -200000
    ax1.text(x, y_rpeak + y_adjust, label_text, ha='center', va='center', bbox=bbox_props)  # ラベルを中央揃えに設定

# 凡例を表示
legend1 = ax1.legend(loc='upper left', bbox_to_anchor=(0, 1))
legend2 = ax2.legend(loc='upper left', bbox_to_anchor=(0, 0.95))
    
# ラベルを設定
# plt.title('Yearly supercomputer performance')
plt.xlabel('Year')
plt.xticks(rotation=45)  # X軸のラベルを回転させる
plt.tight_layout()
plt.savefig('supercomputer_performance_plot_with_core.png', dpi=600)
plt.show()
```
